## **3. Negative Binomial**


$$
P(X=k)=\binom{k-1}{r-1}p^r(1-p)^{k-r}
$$

$$
\mathbb{E}[X]=\frac{r}{p},\quad
\mathrm{Var}(X)=\frac{r(1-p)}{p^2}
$$


## **4. Geometric**


$$P(X=k)=(1-p)^{k-1}p,\quad k=1,2,\dots$$


$$
\mathbb{E}[X]=\frac{1}{p},\quad
\mathrm{Var}(X)=\frac{1-p}{p^2}
$$

## **4. Exponential**
