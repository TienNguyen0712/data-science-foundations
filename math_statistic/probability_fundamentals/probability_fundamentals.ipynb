- **Kh√¥ng gian m·∫´u (Œ©):** l√† t·∫≠p howpk c·ªßa t·∫•t c·∫£ c√°c k·∫øt qu·∫£ c√≥ th·ªÉ x·∫£y ra c·ªßa ph√©p th·ª≠

- **Bi·∫øn c·ªë (s·ª± ki·ªán):** l√† t·∫≠p con c·ªßa kh√¥ng gian m·∫´u k√Ω hi·ªáu th∆∞·ªùng l√† nh∆∞ng ch·ª© $$ A, B, C $$ ƒë∆∞·ª£c vi·∫øt hoa. Ta c√≥ bi·∫øn c·ªë ch·∫Øc ch·∫Øn th∆∞·ªùng l√† $$P = 1$$ kh√¥ng ch·∫Øc ch·∫Øn l√† $$P = 0$$ v√† tr∆∞·ªùng h·ª£p ng·∫´u nhi√™n s·∫Ω l√† $$0 < P < 1$$

- **X√°c su·∫•t c·ªßa bi·∫øn c·ªë:** D√πng ƒë·ªÉ ƒëo m·ª©c ƒë·ªô x·∫£y ra c·ªßa bi·∫øn c·ªë ƒë√≥

**C√¥ng th·ª©c:**

V·ªõi P(A) l√† x√°c su·∫•t ƒë·ªÉ x·∫£y ra c·ªßa bi·∫øn c·ªë A, n(A) l√† s·ªë l·∫ßn x·∫£y ra c·ªßa bi·∫øn c·ªë A v√† n(Œ©) s·∫Ω l√† kh√¥ng gian m·∫´u

**T√≠nh ch·∫•t**

$$0 < P(A) < 1, P(ùõÄ) = 1$$

L·∫•y v√≠ d·ª• gieo m·ªôt con x√∫c x·∫Øc quan s√°t kh√¥ng gian m·∫´u v·ªõi **Ng·ª≠a (1), X·∫•p (0)**

## **2. Bi·∫øn ng·∫´u nhi√™n**

## **3. K·ª≥ v·ªçng & Ph∆∞∆°ng sai**

**K·ª≥ v·ªçng:** L√† gi√° tr·ªã trung b√¨nh d√†i h·∫°n

- **R·ªùi r·∫°c**
$$\mathbb{E}[X] = \sum_i x_i P(X=x_i)
$$

- **Li√™n t·ª•c**
$$\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x)\,dx
$$

**L·∫•y v√≠ d·ª•:**

Trong m·ªôt app b√°n h√†ng online m·ªói ng√†y s·ªë ƒë∆° h√†ng l√† **ng·∫´u nhi√™n**, ph·ª• thu·ªôc v√†o c√°c y·∫øu t·ªë nh∆∞ giao th√¥ng, th·ªùi ti·∫øt, qu·∫£ng c√°o, ...

Ta m√¥ h√¨nh h√≥a nh∆∞ sau:
- X = **doanh thu trong m·ªôt ng√†y**
- X l√† **bi·∫øn ng·∫´u nhi√™n**

G·ªâa s·ª´ t·ª´ d·ªØ li·ªáu l·ªãch s·ª≠ ta ∆∞·ªõc l∆∞·ª£ng ƒë∆∞·ª£c ph√¢n ph·ªëi nh∆∞ sau

$$DoanhThu = 80, 100, 130$$

X√°c su·∫•t l·∫ßn l∆∞·ª£t l√†

$$XacSuat = 0.2, 0.5, 0.3$$

Ta t√≠nh k·ª≥ v·ªçng nh∆∞ sau:

$$\mathbb{E}[X] = 80 . 0.2 + 100 . 0.5 + 130 . 0.3 = 105
$$

**Nh∆∞ v·∫≠y d·ª±a k·ª≥ v·ªçng m√† ta k·∫øt lu·∫≠n:**

_105 tri·ªáu ch√≠nh l√† doanh thu trung b√¨nh d√†i h·∫°n_

```python
revenue = [80, 100, 130]
probability = [0.2, 0.5, 0.3]

mean = sum([revenue[i] * probability[i] for i in range(len(revenue))])

mean
```


**Ph∆∞∆°ng sai (Variance):** M·ª©c ƒë·ªô ph√¢n t√°n

$$\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Bi·ªÉu th·ªã ƒë·ªô r·ªßi ro v√† bi·∫øn ƒë·ªông c·ªßa t·∫≠p d·ªØ li·ªáu

Gi√∫p ƒë√°nh gi√° r·ªßi ro, ph√°t hi·ªán b·∫•t th∆∞·ªùng

Ta th·ª≠ t√≠nh ph∆∞∆°ng sai b·∫±ng c√¥ng th·ª©c

$$\mathrm{Var}(X) = \mathbb{E}[(X - 105)^2] = (80 - 105)^2 . 0.2 + (100 - 105)^2 . 0.5 + (130 - 105)^2 . 0.3 = 325
$$

ƒê·ªô l·ªách chu·∫©n

$$\sigma = \sqrt{325} = 18$$

**Ta c√≥ th·ªÉ k·∫øt lu·∫≠n nh∆∞ sau:**

*Bi·∫øn ƒë·ªông gi·ªØa doanh thu kho·∫£ng +- 18 tri·ªÅu ƒë·ªìng*



```python
variance = sum([(revenue[i] - mean)**2 * probability[i] for i in range(len(revenue))])

variance, variance**0.5
```

## **5. ƒê·ªãnh l√Ω Bayes**

ƒê∆∞·ª£c d·ª±a tr√™n s√°c xu·∫•t c√≥ ƒëi·ªÅu ki·ªán cho r·∫±ng h·ªçc t·ª´ nhi·ªÅu d·ªØ li·ªáu kh√°c nhau ƒë·ªÉ c√≥ th·ªÉ c·∫≠p nh·∫≠t s√°c xu·∫•t

$$P(A \mid B) =
\frac{P(B \mid A) P(A)}{P(B)}$$

M·∫´u s·ªë

$$P(B) = \sum_i P(B \mid A_i) P(A_i)$$

**Trong ƒë√≥:**
- P(A): Prior (Ni·ªÅm tin ban ƒë·∫ßu)
- P(B | A): Likelihood (D·ªØ li·ªáu)
- P(A | B): Posterior (Ni·ªÅm tin sau khi h·ªçc)

**Gi·∫£ s·ª≠**

- A: Email l√† spam
- B: Email c√≥ t·ª´ "free"

Bayes s·∫Ω h·ªèi r·∫±ng n·∫øu th·∫•y "free" th√¨ kh·∫£ nƒÉng l√† spam l√† bao nhi√™u

**V√≠ d·ª•**

Ta c√≥ m·ªôt h·ªá th·ªëng l·ªçc spam:
- T·ª∑ l·ªá spam: P(S) = 0.2
- N·∫øu l√† spam th√¨ b·ªô l·ªçc ƒë√∫ng 90% t·ª©c P(+ | S) = 0.9
- N·∫øu kh√¥ng spam th√¨ b·ªô l·ªçc nh·∫ßm spam 5% P(+ | S^c) = 0.05
  - S: Spam
  - +: ƒê√∫ng

N·∫øu b·ªô l·ªçc b√°o spam, x√°c su·∫•t email th·ª±c s·ª± l√† spam (S | +) l√† bao nhi√™u


```python
p_s = 0.2 # X√°c su·∫•t spam
p_s_plus = 0.9 # X√°c su·∫•t d·ª± ƒëo√°n spam l√† ƒë√∫ng spam
p_s_not_plus = 0.05 # X√°c su·∫•t d·ª± ƒëo√°n spam nh∆∞ng kh√¥ng ph·∫£i spam


p_plus = p_s_plus * p_s + p_s_not_plus * (1 - p_s) # X√°c su·∫•t d·ª± ƒëo√°n ƒë√∫ng

p_plus_s = (p_s_plus * p_s) / p_plus # X√°c su·∫•t d·ª± ƒëo√°n ƒë√∫ng spam

p_plus_s
```

## **6. Lu·∫≠t s·ªë l·ªõn (LLN)**

Trung b√¨nh gi√° tr·ªã m·∫´u

$$\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$$

Ph√°t bi·ªÉu
$$\bar{X}_n \xrightarrow{n \to \infty} \mathbb{E}[X]$$

Khi ta l·∫•y c√†ng nhi·ªÅu d·ªØ li·ªáu trung b√¨nh m·∫´u c√†ng g·∫ßn v·ªõi gi√° tr·ªã th·∫≠t

**V√≠ d·ª•:**

Tung ƒë·ªìng xu:
- 10 l·∫ßn -> c√≥ th·ªÉ l·ªách
- 10000 l·∫ßn -> g·∫ßn ti·∫øn t·ªõi 0.5

Di·ªÖn gi·∫£i To√°n h·ªçc ta g·ªçi X_i = 1 n·∫øu ra ng·ª´a v√† 0 n√™u ra s·∫•p. Ta s·∫Ω t√≠nh trung b√¨nh k·ª≥ v·ªçng

V·ªõi xu c√¥ng b·∫±ng P(X = 1) = 0.5
$${E}[X_i] = 1 . 0.5 + 0.  0.5 = 0.5$$

Lu·∫≠t s·ªë l·ªõn ta ph√°t bi·ªÉu:  

$$\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i \xrightarrow{n \to \infty} \mathbb{E}[X] = 0.5$$

```python
import numpy as np

np.random.seed(0)
n = 100000
X = np.random.binomial(1, 0.5, size=n)
X.mean()
```

## **7. ƒê·ªãnh l√Ω gi·ªõi h·∫°n trung t√¢m (CLT)**

**Ph√°t bi·ªÉu**: Trung b√¨nh c·ªßa b·∫•t k·ª≥ ph√¢n ph·ªëi n√†o c≈©ng ti·∫øn t·ªõi ph√¢n ph·ªëi chu·∫©n khi ƒë·ªß l·ªõn

$$\frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}}
\xrightarrow{d} \mathcal{N}(0,1)$$

**V√≠ d·ª•**

Bi·∫øt m·ªôt bi·∫øn ng·∫´u nhi√™n c√≥ k·ª≥ v·ªçng u = 10, ƒë·ªô l·ªách chu·∫©n l√† 4. K√≠ch th∆∞·ªõc m·∫´u n = 64. X·∫•p x·ªâ P(X > 11) b·∫±ng gi·ªõi h·∫°n trung t√¢m

**√Åp d·ª•ng Gi·ªõi h·∫°n trung t√¢m:**

$$\bar{X} \approx \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)
$$

N√™n:

- Trung b√¨nh: 10
- ƒê·ªô l·ªách chu·∫©n theo c√¥ng th·ª©c = 0.5

Chu·∫©n h√≥a
$$P(\bar{X} > 11) = P(Z > \frac{(11 - 10)}{0.5}) = P(Z > 2)$$
Tra b·∫≥ng chu·∫©n:
$$P(Z > 2) ‚âà 0.0228$$

```python
np.random.seed(0)

mu, sigma, n = 10, 4, 64
trials = 200000
samples = np.random.normal(mu, sigma, size=(trials, n)) # D√πng normal ƒë·ªÉ l√†m nhanh d·ª± ƒëo√°n
xbar = samples.mean(axis=1)
prob_est = np.mean(xbar > 11)
prob_est
```
